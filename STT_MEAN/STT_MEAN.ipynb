0. Imports

```python
import pandas as pd
import numpy as np
import math
```

1. Load csv

```python
df = pd.read_csv('/content/drive/MyDrive/Škola/WM/STT_MEAN/log_jeden_den.csv', ';', usecols=range(1,13))
df.head()
```

2. Identify sessions based on STT_MEAN ( mean of 'length' )

```python
df['STT_MEAN'] = ''
count = 1
time = 0
C = df["length"].mean(skipna = True)

for i, row in df.iterrows():
  if (i != 0):
    if (df.at[i,'userid'] == df.at[i-1,'userid']):
      if not (math.isnan(df.at[i-1, 'length'])):
        time += df.at[i-1, 'length'].astype(int)
      else:
        count += 1
        time = 0
      if (time > C):
        count += 1
        time = 0
    else:
      count += 1
      time = 0
  df.at[i, 'STT_MEAN'] = count

df.tail()
```

3. Save to csv

```python
df.to_csv('Laca_stt_mean.csv', sep=';')
```
