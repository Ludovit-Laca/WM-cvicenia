0. Imports

```python
import pandas as pd
import numpy as np
import math
```

1. Load csv

```python
df = pd.read_csv('/content/drive/MyDrive/Škola/WM/SLength/log_jeden_den.csv', ';', usecols=range(1,15))
df.tail()
```

2. Identify sessions based on estimation - 10 minutes (SLength)

```python
df['SLength'] = ''
count = 1
time = 0
C = 600 # seconds

for i, row in df.iterrows():
  if (i != 0):
    if (df.at[i,'userid'] == df.at[i-1,'userid']):
      if not (math.isnan(df.at[i-1, 'length'])):
        time += df.at[i-1, 'length'].astype(int)
      else:
        count += 1
        time = 0
      if (time > C):
        count += 1
        time = 0
    else:
      count += 1
      time = 0
  df.at[i, 'SLength'] = count

df.tail()
```

3. Save to csv

```python
df.to_csv('Laca_slength.csv', sep=';')
```
