```python
df['length'] = ''

for i, row in df.iterrows():
  if (i < len(df) - 1):
    if (df.at[i,'userid'] == df.at[i+1,'userid']):
      if ((df.at[i+1,'unixtime'] - df.at[i,'unixtime']) < 3600):
        df.at[i,'length'] = df.at[i+1,'unixtime'] - df.at[i,'unixtime']
      
df.head(15)
```
