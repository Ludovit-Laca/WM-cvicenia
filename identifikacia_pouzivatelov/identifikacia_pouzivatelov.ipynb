0. Imports

```python
import pandas as pd
import numpy as np
```

1. Load csv

```python
df = pd.read_csv('/content/drive/MyDrive/Škola/WM/identifikacia_pouzivatelov/log_jeden_den.csv', ';', usecols=range(1,11))
df.head()
```

2. Sort values by **ip**, then **useragent** and **unixtime**

```python
df = df.sort_values(by = ['ip', 'useragent', 'unixtime'])
df.head(5)
```

3. Define column userid from column **ip** and column **useragent**

```python
previousIp = ""
previousUserAgent = ""
previousUserId = 0
for i, row in df.iterrows():
  if (i == 0):
    df.at[i,'userid'] = 0
    previousIp = row['ip']
    previousUserAgent = row['useragent']
  else:
    if ((row['ip'] == previousIp) & (row['useragent'] == previousUserAgent)):
      df.at[i,'userid'] = previousUserId
    else:
      previousUserId += 1
      previousIp = row['ip']
      previousUserAgent = row['useragent']
    df.at[i,'userid'] = previousUserId

df.tail(5)
```

4. Save to csv

```python
df.to_csv('Laca_identifikacia_pouzivatelov.csv', sep=';')
```
