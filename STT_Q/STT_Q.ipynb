0. Imports

```python
import pandas as pd
import numpy as np
import math
```

1. Load csv

```python
df = pd.read_csv('/content/drive/MyDrive/Škola/WM/STT_Q/log_jeden_den.csv', ';', usecols=range(1,14))
df.tail()
```

2. Replace NaN values in 'length' with 0

```python
df['length'].fillna(0, inplace=True)
```

3. Identify sessions based on STT_Q

```python
df['STT_Q'] = ''
count = 1
time = 0
C = np.quantile(df.length,[0.25, 0.75])[1] + (1.5*15)

for i, row in df.iterrows():
  if (i != 0):
    if (df.at[i,'userid'] == df.at[i-1,'userid']):
      if not (math.isnan(df.at[i-1, 'length'])):
        time += df.at[i-1, 'length'].astype(int)
      else:
        count += 1
        time = 0
      if (time > C):
        count += 1
        time = 0
    else:
      count += 1
      time = 0
  df.at[i, 'STT_Q'] = count

df.tail()
```

4. Save to csv

```python
df.to_csv('Laca_stt_q.csv', sep=';')
```
