0.) Imports

```
import numpy as np
import pandas as pd

import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize, sent_tokenize 
from nltk.stem import WordNetLemmatizer 

nltk.download('punkt')
nltk.download('averaged_perceptron_tagger')
nltk.download('stopwords')
nltk.download('wordnet')

stop_words=set(stopwords.words("english"))
```

1.) Load texts

```
text1 = open("/content/drive/MyDrive/Škola/WM/reprezentacia_textu III/text1.txt", "r").read()
text2 = open("/content/drive/MyDrive/Škola/WM/reprezentacia_textu III/text2.txt", "r").read()
text3 = open("/content/drive/MyDrive/Škola/WM/reprezentacia_textu III/text3.txt", "r").read()
text4 = open("/content/drive/MyDrive/Škola/WM/reprezentacia_textu III/text4.txt", "r").read()
text5 = open("/content/drive/MyDrive/Škola/WM/reprezentacia_textu III/text5.txt", "r").read()
```

2.) Break texts into *sentences*

```
# tokenize sentences in texts and add them to array 
text_sent= np.array([sent_tokenize(text1),
                sent_tokenize(text2),
                sent_tokenize(text3),
                sent_tokenize(text4),
                sent_tokenize(text5)], dtype=object)

print(text_sent)
```

3.) Tokenize sentences into words, remove stopwords, lemmazie etc.

```
# initialize lemmatizer 
lemmatizer = WordNetLemmatizer();

#final array
array_words = []

#counter for sentences
sent_counter = 1
for text in text_sent:
  for sentence in text:

    #tokenize sentence into words
    words_sent = word_tokenize(sentence)

    #counter for words in sentences
    word_counter = 1
    for word in words_sent:
      if word not in stop_words:
        pos_tag = nltk.tag.pos_tag([word])
        lemmatizer_word = lemmatizer.lemmatize(word.lower(), "v");
        array_words.append([sent_counter, pos_tag[0][0], pos_tag[0][1], lemmatizer_word, word_counter])
        word_counter += 1
    sent_counter += 1

print("Filterd Sentence:",array_words)
```

4.) Create dataframe

```
df = pd.DataFrame(array_words, columns=['ID_veta', 'Slovo', 'Tag', 'Lemma', 'Poradie'])
df.set_index('ID_veta',inplace=True)
df
```

5.) Export to excel

```
# export to excel
with pd.ExcelWriter('reprezentacia_textu_3.xlsx') as writer:  
    df.to_excel(writer, sheet_name='Hárok1')
```
